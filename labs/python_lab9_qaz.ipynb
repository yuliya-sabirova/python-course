### 6-тапсырма: САТЖ шешу (2 балл)

**Міндет:** Нұсқаңызға сәйкес екі сызықтық алгебралық теңдеулер жүйесін (САТЖ) шешіңіз. Әр шешім үшін тексеруді міндетті түрде орындаңыз.

**1-нұсқа:**
```
1) 2x₁ - 3x₂ - x₃ = 3      2) x₁ - x₂ - x₃ + x₄ = 0
   3x₁ + x₂ + 2x₃ = 1         x₁ + x₂ + x₃ - x₄ = 4
   x₁ + 4x₂ - 2x₃ = -6        x₁ + x₂ - x₃ + x₄ = 6
                             x₁ - x₂ + x₃ + x₄ = 2
```

**2-нұсқа:**
```
1) x₁ + 2x₂ + 3x₃ = 6      2) 2x₁ + x₂ + x₃ + x₄ = 1
   4x₁ + 5x₂ + 6x₃ = 9         x₁ + 2x₂ + x₃ + x₄ = 2
   7x₁ + 8x₂ = -6             x₁ + x₂ + 2x₃ + x₄ = 3
                             x₁ + x₂ + x₃ + 2x₄ = 4
```

**3-нұсқа:**
```
1) x₁ + 2x₂ + 3x₃ = 1      2) 2x₁ + x₂ + x₃ + x₄ = 2
   x₂ + 2x₃ + 3x₄ = 2         x₁ + 2x₂ + x₃ + x₄ = 1
   x₃ + 2x₄ = 3             x₁ + x₂ + 2x₃ + x₄ = 2
   x₁ + x₂ + x₃ + x₄ = 4        x₁ + x₂ + x₃ + 2x₄ = 1
```

**4-нұсқа:**
```
1) x₁ + 2x₂ - x₃ = 2       2) x₁ + 2x₂ + 3x₃ - 2x₄ = 6
   2x₁ - x₂ + x₃ = 3         2x₁ - x₂ - 2x₃ - 3x₄ = 8
   x₂ + 2x₃ = -2             3x₁ + 2x₂ - x₃ + 2x₄ = 4
                             2x₁ - 3x₂ + 2x₃ + x₄ = -8
```

### 7-тапсырма: Матрицалық теңдеуді шешу (2 балл)

**Міндет:** Нұсқаңызға сәйкес матрицалық теңдеуді шешіп, тексеру орындаңыз.

-   **1-нұсқа:** `AX = B`
    `A = np.array([[1, 2], [3, -1]])`, `B = np.array([[4, -1], [2, 5]])`

-   **2-нұсқа:** `XA = B`
    `A = np.array([[1, 2, -1], [2, -1, 1], [-1, 2, 1]])`, `B = np.array([[1, -2, 1], [2, -1, 1], [-1, 2, 1]])`

-   **3-нұсқа:** `AX = B`
    `A = np.array([[3, -1], [5, -2]])`, `B = np.array([[14, 16], [9, 10]])`

-   **4-нұсқа:** `XA = B`
    `A = np.array([[5, 3, 1], [1, -3, -2], [-5, 2, 1]])`, `B = np.array([[-8, 3, 0], [-5, 9, 0], [-2, 15, 0]])`

> **Кеңес:** `XA = B` теңдеуі үшін шешім `X = B @ A⁻¹` болады. `AX = B` үшін шешім `X = A⁻¹ @ B`. Көбейту реті маңызды!
