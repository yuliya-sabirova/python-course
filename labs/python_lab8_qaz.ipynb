## 2-бөлім. "Геометриялық фигура" класы (6 балл)

**Мәселе:** Нұсқаңызға сәйкес геометриялық фигураны сипаттау үшін үш класстың бірін іске асыру.

**Нұсқаны анықтау:** Сіздің нұсқаңыз `(N - 1) % 3 + 1` формуласы бойынша анықталады, мұндағы `N` — топ тізіміндегі сіздің нөміріңіз.
-   **1-нұсқа:** `Triangle` класы
-   **2-нұсқа:** `Rectangle` класы
-   **3-нұсқа:** `Circle` класы

### Класстарға қойылатын талаптар:

#### 1-нұсқа: `Triangle`
-   **Атрибуттар:** `side_a`, `side_b`, `side_c` (үшбұрыштың қабырғалары).
-   **Әдістер:**
    -   Конструктор `__init__`.
    -   Әр қабырға үшін геттерлер мен сеттерлер. Класс атрибуттарына қол жеткізу деңгейлерімен тәжірибе жасаңыз.
    -   `area()`: үшбұрыштың ауданын есептейді және қайтарады (Герон формуласын қолдануға болады).
    -   `perimeter()`: периметрін есептейді және қайтарады.
    -   `check_type()`: үшбұрыштың тең бүйірлі, тең қабырғалы немесе әртүрлі қабырғалы екенін тексереді және шығарады.
-   **Демонстрацияға арналған бағдарлама:** Класс объектісін құрып, барлық әдістерді шақырыңыз және нәтижелерді экранға шығарыңыз.

#### 2-нұсқа: `Rectangle`
-   **Атрибуттар:** `side_a`, `side_b` (тіктөртбұрыштың қабырғалары).
-   **Әдістер:**
    -   Конструктор `__init__`.
    -   Әр қабырға үшін геттерлер мен сеттерлер. Класс атрибуттарына қол жеткізу деңгейлерімен тәжірибе жасаңыз.
    -   `area()`: ауданын есептейді және қайтарады.
    -   `perimeter()`: периметрін есептейді және қайтарады.
    -   `is_square()`: егер тіктөртбұрыш шаршы болса `True`, әйтпесе `False` қайтарады.
-   **Демонстрацияға арналған бағдарлама:** Класс объектісін құрып, барлық әдістерді шақырыңыз және нәтижелерді экранға шығарыңыз.
