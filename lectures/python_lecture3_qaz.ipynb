### **1-бөлім: Тізімдер (Lists) — Мүмкіндігі мол коллекциялар**

Елестетіңіз, сізге бір мәнді емес, бірнеше мәнді сақтау керек: мысалы, сауда тізімі, семестрдегі бағалар немесе топтағы студенттердің есімдері. Ол үшін Python-да **тізімдер** қолданылады.

**Тізім (list)** — бұл реттелген және өзгертілетін элементтер жиынтығы.

*   **Реттелген:** Элементтердің орналасу тәртібі маңызды, және егер сіз оны өзіңіз өзгертпесеңіз, ол өзгермейді.
*   **Өзгертілетін:** Тізімді құрғаннан кейін элементтерді қосуға, жоюға немесе өзгертуге болады.
*   **Әртүрлі деректер түрлерін қамтуы мүмкін:** Бір тізімде сандар, жолдар, тіпті басқа тізімдер де болуы мүмкін.

#### **1.1. Тізімді құру және элементтерге қол жеткізу**

Тізім төртбұрышты жақшалар `[]` арқылы құрылады, ал элементтер үтір арқылы жазылады.

````python
# Біздің алғашқы тізімді құрайық
planets = ["Меркурий", "Шолпан", "Жер", "Марс"]
print(planets)
````

Элементке қол жеткізу үшін оның төртбұрышты жақшадағы **индексін** (реттік нөмірін) көрсету керек. **Маңызды:** Python-да нөмірлеу нөлден басталады!

````python
# Бірінші элементті алайық (индекс 0)
first_planet = planets[0]
print(f"Күннен бастап бірінші ғаламшар: {first_planet}")

# Үшінші элементті алайық (индекс 2)
our_planet = planets[2]
print(f"Біздің ғаламшар: {our_planet}")
````

Теріс индекстерді де қолдануға болады. `[-1]` — бұл соңғы элемент, `[-2]` — соңғының алдындағы элемент және т.б.

````python
# Соңғы элементті алайық
last_planet = planets[-1]
print(f"Тізімдегі соңғы ғаламшар: {last_planet}")
````

#### **1.2. Тізімдердің негізгі әдістері**

Тізімдердің оларды басқаруға мүмкіндік беретін кіріктірілген "командалары" (әдістері) бар. Олардың негізгілерін мысал арқылы қарастырайық.

````python
# 1. Бастапқы тізімді құрамыз
tools = ["балға", "бұрауыш", "рулетка", "бұрауыш"]
print(f"1. Бастапқы тізім: {tools}")

# 2. append(): элементті тізімнің СОҢЫНА қосады
tools.append("дрель")
print(f"2. append('дрель') әдісінен кейін: {tools}")

# 3. insert(): элементті көрсетілген позицияға қосады (индекс, мән)
tools.insert(1, "қысқыш")
print(f"3. insert(1, 'қысқыш') әдісінен кейін: {tools}")

# 4. remove(): элементтің мәні бойынша БІРІНШІ кездескенін жояды
tools.remove("бұрауыш")
print(f"4. remove('бұрауыш') әдісінен кейін: {tools}")

# 5. pop(): элементті индексі бойынша жояды және оның мәнін ҚАЙТАРАДЫ
removed_tool = tools.pop(2)
print(f"5. 2-индекс бойынша жойылған элемент: {removed_tool}")
print(f"   pop(2) әдісінен кейінгі тізім: {tools}")

# 6. index(): элементтің БІРІНШІ кездесуінің индексін қайтарады
index_of_hammer = tools.index("балға")
print(f"6. 'балға' элементінің индексі: {index_of_hammer}")

# 7. count(): элементтің тізімде қанша рет кездесетінін санайды
screwdriver_count = tools.count("бұрауыш")
print(f"7. Тізімдегі бұрауыштар саны: {screwdriver_count}")

# 8. copy(): тізімнің ТОЛЫҚ (тәуелсіз) көшірмесін жасайды
tools_copy = tools.copy()
print(f"8. Түпнұсқа: {tools}, Көшірме: {tools_copy}")
# Түпнұсқаның өзгермегеніне көз жеткізу үшін көшірмені өзгертеміз
tools_copy.append('жаңа жазба')
print(f"   Көшірмені өзгерткеннен кейін -> Түпнұсқа: {tools}, Көшірме: {tools_copy}")

# 9. clear(): тізімнен БАРЛЫҚ элементтерді жояды
tools.clear()
print(f"9. clear() әдісінен кейінгі бастапқы тізім: {tools}")
````

#### **1.3. Кірістірілген тізімдер және кесінділер**

**Кірістірілген тізімдер** — бұл басқа тізімдердің ішінде орналасқан тізімдер. Олар матрицалар немесе күрделі деректер құрылымдарын жасау үшін пайдалы.

````python
# Кірістірілген тізім мысалы: студенттер туралы деректер
# [Аты-жөні, ID, [бағалары]]
student_data = [
    ["Иванов", 101, [5, 4, 5, 5]],
    ["Петров", 102, [3, 4, 4, 5]],
    ["Сидоров", 103, [5, 5, 5, 5]]
]
print(student_data)
````

Элементтерге қол жеткізу үшін қос индекстеу қолданылады: `тізім[жол_индексі][баған_индексі]`.

````python
# Екінші студенттің атын алайық (индекс 1, элемент 0)
petrov_name = student_data[1][0]
print(f"Екінші студенттің аты-жөні: {petrov_name}")

# Сидоровтың бағаларын алайық (индекс 2, элемент 2)
sidorov_grades = student_data[2][2]
print(f"Сидоровтың бағалары: {sidorov_grades}")

# Ал енді Сидоровтың бірінші бағасын алайық
sidorov_first_grade = student_data[2][2][0]
print(f"Сидоровтың бірінші бағасы: {sidorov_first_grade}")
````

**Кесінділер (Slices)** бір элементті емес, бірден тұтас ауқымды алуға мүмкіндік береді. Синтаксис: `тізім[start:stop:step]`.

*   `start`: басталатын индекс (оны қоса алғанда).
*   `stop`: аяқталатын индекс (оны ҚОСПАЙ).
*   `step`: қадам.

````python
temperatures = [21, 23, 25, 24, 22, 26, 20]
print(f"Температуралардың бастапқы тізімі: {temperatures}")

# 1. Алғашқы екі күннің температурасы (0-ден 2-ге дейінгі индекс, 2-ні қоспай)
first_two_days = temperatures[:2]
print(f"Алғашқы екі күн: {first_two_days}")

# 2. 3-тен 5-күнге дейінгі температуралар (индекстер 2, 3, 4)
mid_week = temperatures[2:5]
print(f"Апта ортасы: {mid_week}")

# 3. Күн ара температуралар (біріншіден бастап, қадам 2)
every_other_day = temperatures[::2]
print(f"Күн ара температуралар: {every_other_day}")

# 4. Температуралар кері тәртіпте (қадам -1)
reversed_temps = temperatures[::-1]
print(f"Кері тәртіптегі тізім: {reversed_temps}")
````

### **2-бөлім: Циклдер (Loops) — Бірдей әрекеттерді автоматтандыру**

Циклдер бірдей код блогын бірнеше рет орындау үшін қажет. Бұл бізді бір типті жолдарды көп рет жазудан құтқарады.

#### **2.1. `for` циклі**

`for` циклі қандай да бір тізбектің (тізім, жол, сандар диапазоны) әрбір элементін "аралап" шыққымыз келгенде өте ыңғайлы.

**Синтаксис:**
```python
for айнымалы in тізбек:
    # Қайталау керек әрекеттер
    # Бұл код блогы "цикл денесі" деп аталады
```

**1-мысал: Тізім бойынша итерация**

````python
# Әрбір сөзді бас әріптерге айналдыру
words = ["hello", "world", "python"]

for word in words:
    print(word.upper())
````

**2-мысал: `range()` функциясын қолдану**

`range()` функциясы итерация жасауға өте ыңғайлы сандар тізбегін құрайды.

````python
# 0-ден 50-ге дейінгі 5-ке еселік барлық сандарды шығару
# range(start, stop, step)
for i in range(0, 51, 5):
    print(i)
````

#### **2.2. Есептеулері бар цикл**

Циклдерде жиі қандай да бір нәтижені жинақтау қажет болады (қосынды, көбейтінді және т.б.). Ол үшін цикл алдында **аккумулятор**-айнымалы құрылады, ол әр итерацияда жаңартылып отырады.

````python
# Себеттегі тауарлардың жалпы құнын есептеу
prices = [120, 55.5, 30, 150.25]
total_cost = 0  # 1. Аккумуляторды инициализациялаймыз

for price in prices:
    total_cost = total_cost + price  # 2. Оны циклде жаңартамыз

print(f"Жалпы құны: {total_cost:.2f} тг.") # 3. Нәтижені циклден кейін қолданамыз
````

#### **2.3. `if` шартты операторы бар `for` циклі**

Бұл элементтерді талдауға және белгілі бір шартқа сәйкес келетіндерін ғана таңдап алуға мүмкіндік беретін қуатты комбинация.

**Тапсырма:** Сөздер тізімінен ұзындығы 5 символдан асатын сөздерден ғана тұратын жаңа тізім құру.

````python
all_words = ["algorithm", "data", "science", "code", "python", "learning"]
long_words = [] # 1. Нәтижелер үшін бос тізім құрамыз

for word in all_words:
    # 2. ӘРБІР элемент үшін шартты тексереміз
    if len(word) > 5:
        # 3. Егер шарт ақиқат болса, элементті жаңа тізімге қосамыз
        long_words.append(word)

print(f"Бастапқы тізім: {all_words}")
print(f"Ұзындығы 5 символдан асатын сөздер: {long_words}")
````

#### **2.4. Кірістірілген циклдер**

Бұл басқа циклдің ішіндегі цикл. Сыртқы цикл бір итерацияны орындайды, ал ішкі цикл осы уақыт ішінде ТОЛЫҒЫМЕН орындалып үлгереді.

**Мысал: Координаттар торын құру**

#### **2.5. Циклді басқару: break және continue**

Кейде циклдің стандартты жұмысы жеткіліксіз болады. Бізге циклді мерзімінен бұрын тоқтату немесе ағымдағы қадамды өткізіп, келесісіне көшу қажет болуы мүмкін. Ол үшін `break` және `continue` операторлары бар.

**break операторы (Мерзімінен бұрын шығу)**

break операторы өзі орналасқан ең ішкі циклді **дереу және толығымен аяқтайды**. Бағдарламаның орындалуы циклден *кейінгі* келесі код жолынан жалғасады.

**Аналогия:** Елестетіңіз, сіз сөреден белгілі бір кітапты іздеп жатырсыз. Сіз оларды бірінен соң бірін қарайсыз. Қажетті кітапты тапқаннан кейін, сіз іздеуді тоқтатып, оны аласыз. Сіз сөреде қалған кітаптарды қарамайсыз. `break` — бұл "табу және іздеуді тоқтату" әрекеті.
Әдетте `break` `if` шартты операторының ішінде қолданылады.

**Мысал:** Тізімнен "admin" аты бар бірінші пайдаланушыны тауып, іздеуді тоқтату.

````python
users = ["anna", "ivan", "peter", "admin", "olga", "sergey"]

print("Әкімшіні іздеуді бастаймыз...")

for user in users:
    print(f"Пайдаланушыны тексеру: {user}")
    if user == "admin":
        print("Әкімші табылды!")
        # Мақсатқа жеткендіктен, циклден дереу шығамыз
        break

print("Іздеу аяқталды.") 
# Бұл код break-тен кейін бірден орындалады
````

Көріп отырғаныңыздай, "admin" табылғаннан кейін "olga" және "sergey" пайдаланушылары тексерілмеді де.

**continue операторы (Итерацияны өткізіп жіберу)**

`continue` операторы **ағымдағы итерациядағы кодтың қалған бөлігін өткізіп жібереді** де, дереу циклдің келесі итерациясының басына өтеді. Бұл кезде цикл аяқталмайды.

**Аналогия:** Елестетіңіз, сіз құжаттар бумасын өңдеп жатырсыз және сізге барлық бос бланкілерді шетке қою керек. Сіз құжатты аласыз, оның бос екенін көресіз, оны шетке қоясыз (`continue`) және бірден келесісін аласыз. Сіз бүкіл буманы өңдеуді тоқтатпайсыз (`break`), тек бір нақты құжатты өткізіп жібересіз.

**Мысал:** Тізімдегі теріс сандар мен нөлді елемей, тек оң сандардың қосындысын есептеу.

````python
numbers = [10, -5, 20, 0, -15, 30]
positive_sum = 0

print("Тек оң сандардың қосындысын есептейміз...")

for num in numbers:
    # Егер сан оң болмаса...
    if num <= 0:
        print(f"Санды өткізіп жібереміз: {num}")
        # ...осы итерациядағы қалған кодты өткізіп, келесі санға көшеміз
        continue
    
    # Бұл код тек оң сандар үшін орындалады
    print(f"Санды қосамыз: {num}")
    positive_sum += num

print(f"\nОң сандардың қосындысы: {positive_sum}")
````

Назар аударыңыз: цикл -5, 0 немесе -15-ке кездескенде, ол `continue` орындап, `positive_sum += num` жолына жетпеді.

**Негізгі айырмашылық:**
`break:` "Жетеді, бүкіл циклден шығамыз".
`continue:` "Бұл элементпен болдық, келесісін бер".

### **3-бөлім: `while` циклі**

`while` ("әзірше") циклі белгілі бір шарт **ақиқат болғанша** код блогын орындайды. Ол қанша итерация қажет екенін алдын-ала білмеген кезде өте ыңғайлы.

**Синтаксис:**
```python
while шарт:
    # Цикл денесі
    # МАҢЫЗДЫ: мұнда шартқа әсер ететін нәрсе болуы керек,
    # әйтпесе цикл шексіз болып кетеді!
```

**Мысал: Тоқтату-сөзіне дейін деректер жинау**

````python
user_inputs = []
current_input = ""

print("Сөздерді енгізіңіз. Аяқтау үшін 'stop' деп теріңіз.")

while current_input != "stop":
    current_input = input("> ")
    if current_input != "stop":
        user_inputs.append(current_input)

print(f"Сіз келесі сөздерді енгіздіңіз: {user_inputs}")
````

#### **3.1. `for` циклін `while` цикліне ауыстыру**

Кез-келген `for` циклін `while` арқылы қайта жазуға болады.

**Тапсырма:** Тізімнің барлық элементтерін шығару.

````python
fruits = ["алма", "банан", "шие"]

# FOR арқылы жүзеге асыру
print("for арқылы шығару:")
for fruit in fruits:
    print(fruit)

# Сол тапсырма WHILE арқылы
print("\nwhile арқылы шығару:")
index = 0 # 1. Индекс-санауыш қажет
while index < len(fruits): # 2. Шарт: индекс тізім ұзындығынан кіші болғанша
    print(fruits[index]) # 3. Элементті индекс бойынша қолданамыз
    index += 1 # 4. Индексті МІНДЕТТІ ТҮРДЕ арттырамыз!
````

#### **3.2. Практикалық мысал: Қарапайым банк терминалының симуляторы**

````python
balance = 1000
is_running = True

print("Біздің банкоматқа қош келдіңіз!")

# is_running айнымалысы True болғанша цикл жұмыс істейді
while is_running:
    print("\n--------------------")
    print("Қолжетімді операциялар:")
    print("1 - Балансты көрсету")
    print("2 - Қаражат салу")
    print("3 - Қаражатты шешіп алу")
    print("4 - Шығу")
    
    choice = input("Операцияны таңдаңыз (1-4): ")

    if choice == '1':
        # Тек ағымдағы баланс мәнін шығарамыз
        print(f"\nСіздің ағымдағы балансыңыз: {balance} тг.")
    
    elif choice == '2':
        # Соманы сұрап, оны бүтін санға АЙНАЛДЫРАМЫЗ
        amount_str = input("Салатын соманы енгізіңіз: ")
        amount = int(amount_str) # Назар аударыңыз: егер сан емес нәрсе енгізілсе, бағдарлама қатемен аяқталады!
        
        if amount > 0:
            balance += amount
            print(f"Сіз {amount} тг. сәтті салдыңыз. Жаңа баланс: {balance} тг.")
        else:
            # Теріс соманы енгізбеу үшін қарапайым шарт
            print("Қате: салынатын сома оң болуы керек.")

    elif choice == '3':
        amount_str = input("Шешетін соманы енгізіңіз: ")
        amount = int(amount_str) # Назар аударыңыз: егер сан емес нәрсе енгізілсе, бағдарлама қатемен аяқталады!

        if amount <= 0:
            # Соманың оң екенін тексереміз
            print("Қате: шешілетін сома оң болуы керек.")
        elif amount > balance:
            # Шотта қаражаттың жеткілікті екенін тексереміз
            print("Қате: шотта қаражат жеткіліксіз.")
        else:
            # Егер барлық тексерулерден өтсе, ақшаны шешеміз
            balance -= amount
            print(f"Сіз {amount} тг. сәтті шештіңіз. Жаңа баланс: {balance} тг.")
            
    elif choice == '4':
        print("\nБіздің банкоматты пайдаланғаныңызға рахмет. Сау болыңыз!")
        is_running = False  # Циклді аяқтау үшін айнымалының мәнін өзгертеміз
        
    else:
        # Бұл блок пайдаланушы '1', '2', '3' немесе '4'-тен басқа нәрсе енгізсе іске қосылады
        print("\nҚате таңдау. 1-ден 4-ке дейінгі операцияны таңдаңыз.")
````

### **4-бөлім: Пайдалы әдістер және қателерді өңдеу**

#### **4.1. Циклдегі типтерді түрлендіру**

Деректер жиі сандар болса да, жолдар түрінде келеді. Мысалы, `['5', '12', '3']`. Оларды қосуға болмайды ("5" + "12" = "512"). Алдымен оларды түрлендіру керек.

````python
str_numbers = ['5', '12', '3', '100']
int_numbers = []
total_sum = 0

for s_num in str_numbers:
    num = int(s_num) # 1. Жолды санға түрлендіреміз
    int_numbers.append(num) # Жаңа тізімге қосамыз (енді сан ретінде)
    total_sum += num # Қосамыз

print(f"Сандар тізімі (int типі): {int_numbers}")
print(f"Олардың қосындысы: {total_sum}")
````

#### **4.2. Ерекше жағдайларды өңдеу: `try-except`**

Егер жоғарыдағы кодта пайдаланушы санның орнына "сәлем" сөзін енгізсе не болады? Бағдарлама `ValueError` қатесімен "құлайды". Мұны болдырмауға болады.

**Ерекше жағдай (Exception)** — бұл бағдарламаның орындалуы кезінде пайда болатын қате.

`try-except` конструкциясы мұндай қатені "ұстап алуға" және бағдарламаның барлық жұмысын тоқтатпай, балама кодты орындауға мүмкіндік береді.

**Бұл не үшін қажет?**
Бағдарламаларды сенімді және дұрыс емес енгізуден немесе күтпеген жағдайлардан қорғалған ету үшін.

**Қалай қолдануға болады?**
1.  `try` блогына қате тудыруы мүмкін "қауіпті" код орналастырылады.
2.  `except` блогына, **ЕГЕР** `try` блогында қате орын алса, орындалатын код орналастырылады.

**Қарапайым мысал:**

````python
user_input = input("Сан енгізіңіз: ")

try:
    # Енгізілгенді санға түрлендіруге тырысамыз
    number = int(user_input)
    print(f"Сіз {number} санын енгіздіңіз. Оның квадраты: {number**2}")
except ValueError:
    # Бұл код тек int() жолды түрлендіре алмаған жағдайда ғана іске қосылады
    print("Қате! Сіз сан емес, мәтін енгіздіңіз.")

print("Бағдарлама жұмысын жалғастыруда...")
````

**Пайдаланушы енгізуін тексеру (валидация)**

Енді сенімді деректер сұрауын жасау үшін `while`, `try-except` және `if` операторларын біріктірейік.

**Тапсырма:** Пайдаланушыдан (1-ден 5-ке дейін) баға сұрау, ол осы ауқымда дұрыс бүтін санды енгізгенше жалғастыру.

````python
# while True шексіз цикл құрады. Одан тек break командасымен шығуға болады.
while True:
    rating_input = input("Өтінемін, бағаңызды енгізіңіз (1-ден 5-ке дейін): ")

    try:
        # 1. Енгізілгенді санға түрлендіруге тырысамыз
        rating = int(rating_input)

        # 2. Егер сәтті болса, диапазонды тексереміз
        if 1 <= rating <= 5:
            # 3. Егер барлық тексерулерден өтсе - циклден шығамыз
            print(f"Рахмет! Сіздің бағаңыз: {rating}")
            break
        else:
            # Сан енгізілді, бірақ ол қажетті диапазонда емес
            print("Қате: баға 1-ден 5-ке дейінгі диапазонда болуы керек.")

    except ValueError:
        # 4. Егер int(rating_input) қате тудырса, бұл туралы хабарлаймыз
        print("Қате: өтінемін, тек бүтін сандарды енгізіңіз.")
````
