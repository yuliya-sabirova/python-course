# №5 Дәріс: Функциялар және кодты құрылымдау

### Дәріс мақсаттары:
1.  **Монолитті код мәселесін түсіну:** Неліктен барлық кодты бір үлкен скриптте жазу жаман тәжірибе екенін түсіну.
2.  **Функцияларды құруды меңгеру:** `def` көмегімен жеке функцияларды анықтауды, оларға деректерді (аргументтерді) беруді және нәтижені (`return`) алуды үйрену.
3.  **Көріну аймақтарын зерттеу:** Жергілікті және жаһандық айнымалылар арасындағы айырмашылықты түсіну.
4.  **Декомпозицияны үйрену:** Күрделі тапсырмаларды қарапайым және басқарылатын кіші тапсырмаларға бөлу үшін функцияларды қолдану.

## 1-бөлім. Скрипттен бағдарламаға дейін: Функциялар не үшін қажет?

Осы уақытқа дейін біз кодты бірінен соң бірі, жоғарыдан төмен қарай орындалатын етіп жаздық. Бұл **монолитті скрипт** деп аталады. Қарапайым тапсырмалар үшін бұл жұмыс істейді, бірақ бағдарламаның әртүрлі бөліктерінде бірдей әрекетті орындау қажет болса ше? Мысалы, тіктөртбұрыштың ауданын есептеу.

**Мәселе: Кодты қайталау (DRY қағидасы)**

```python
# Бірінші тіктөртбұрышқа арналған деректер
length1 = 10
width1 = 5
area1 = length1 * width1
print(f"Бірінші тіктөртбұрыштың ауданы: {area1}")

# Кодтың басқа жерінде...

# Екінші тіктөртбұрышқа арналған деректер
length2 = 7
width2 = 3
area2 = length2 * width2
print(f"Екінші тіктөртбұрыштың ауданы: {area2}")
```

Біз бірдей логиканы екі рет жаздық. Егер формуланы өзгерткіміз келсе (мысалы, коэффициент қосу), оны **әрбір** жерде өзгертуге тура келеді. Бұл ыңғайсыз және қателіктерге әкеледі.

Бағдарламалауда **DRY — Don't Repeat Yourself** (Өзіңді қайталама) деген маңызды қағида бар. Функциялар — осы қағиданы ұстанудың негізгі құралы.

**Функция** — бұл белгілі бір тапсырманы орындайтын, атауы бар, қайта пайдалануға болатын код блогы.

## 2-бөлім. Функцияның анатомиясы

### 2.1. Функцияны анықтау және шақыру

Функция `def` кілт сөзі арқылы құрылады (анықталады).

```python
# 1. Функцияны анықтау
def say_hello():
    # Функция денесі (шегініспен жазылған код)
    print("Сәлем, әлем!")
    print("Бұл менің бірінші функциям.")

# 2. Функцияны шақыру
print("Бағдарлама басталды.")
say_hello() # Функция ішіндегі код осы жерде орындалады
print("Бағдарлама жұмысын жалғастыруда...")
say_hello() # Біз оны қалағанымызша шақыра аламыз
print("Бағдарлама аяқталды.")
```

### 2.2. Параметрлер және аргументтер: функцияға деректерді беру

Функциялар жұмыс істеу үшін кіріс деректерін қабылдай алады.
*   **Параметр** — бұл функцияны *анықтау* кезінде жақша ішінде көрсетілген айнымалы. Бұл деректерге арналған "үлгі".
*   **Аргумент** — бұл функцияны *шақыру* кезінде берілетін нақты мән.

```python
# Мұнда 'name' және 'age' — параметрлер
def print_user_info(name, age):
    print(f"Пайдаланушы аты: {name}")
    print(f"Жасы: {age} жаста")

# Мұнда "Алиса" және 21 — аргументтер
print_user_info("Алиса", 21)

# Ал мұнда "Борис" және 35
print_user_info("Борис", 35)
```

### 2.3. Мәнді қайтару: `return`

Көбінесе функция бірдеңені басып шығарып қана қоймай, оны әрі қарай пайдалану үшін өз жұмысының **нәтижесін қайтаруы** керек. Ол үшін `return` операторы қолданылады.

Ауданды есептеу мысалымызды қайта жазайық.

```python
def calculate_area(length, width):
    area = length * width
    return area # Функция area айнымалысының мәнін қайтарады

# Функцияны шақырып, нәтижесін айнымалыға сақтаймыз
area1 = calculate_area(10, 5)
area2 = calculate_area(7, 3)

print(f"Біріншісінің ауданы: {area1}")
print(f"Екіншісінің ауданы: {area2}")
print(f"Жалпы ауданы: {area1 + area2}") # Енді біз нәтижемен жұмыс істей аламыз!
```

**`return` туралы маңызды жайттар:**
*   `return` орындалған кезде, функцияның жұмысы **дереу тоқтатылады**.
*   Функция кез келген деректерді қайтара алады: сандар, жолдар, тізімдер, сөздіктер, тіпті `None`.
*   Егер функцияда `return` болмаса, ол жанама түрде `None` қайтарады.

```python
def check_number(n):
    if n > 0:
        return "Оң сан"
        print("Бұл код ешқашан орындалмайды")
    elif n < 0:
        return "Теріс сан"
    else:
        return "Нөл"

result = check_number(-5)
print(result)
```

## 3-бөлім. Аргументтердің кеңейтілген мүмкіндіктері

### 3.1. Позициялық және атаулы аргументтер

*   **Позициялық:** Мәндер параметрлер анықталған ретпен беріледі.
*   **Атаулы (кілтті):** Мәндер параметрдің атын көрсету арқылы беріледі (`аты=мән`), бұл жағдайда реті маңызды емес.

```python
def create_user(login, password, role):
    print(f"{login} пайдаланушысы {role} рөлімен құрылды")

# Позициялық аргументтер
create_user("admin", "12345", "administrator")

# Атаулы аргументтер (реті маңызды емес)
create_user(role="guest", login="alice", password="qwerty")

# Араластыруға болады, бірақ позициялықтар әрқашан бірінші келеді
create_user("bob", role="user", password="pass")
```

### 3.2. Әдепкі аргументтер

Параметрге әдепкі мән беруге болады. Егер функцияны шақыру кезінде бұл параметрге аргумент берілмесе, әдепкі мән қолданылады. **Әдепкі мәндері бар параметрлер әдеттегі параметрлерден кейін келуі керек.**

```python
# 'role' параметрінің әдепкі мәні "user"
def create_user_v2(login, password, role="user"):
    print(f"{login} пайдаланушысы {role} рөлімен құрылды")

# Рөлді көрсетпей шақыру, әдепкі мән қолданылады
create_user_v2("charlie", "zxcvb")

# Рөлді нақты көрсетіп шақыру
create_user_v2("dave", "asdfg", role="moderator")
```

## 4-бөлім. Айнымалылардың көріну аймағы

**Көріну аймағы (scope)** — бұл айнымалы қолжетімді болатын бағдарлама бөлігі.

*   **Жаһандық көріну аймағы:** Бағдарламаның негізгі бөлігінде анықталған айнымалылар. Олар барлық жерде қолжетімді.
*   **Жергілікті көріну аймағы:** Функцияның **ішінде** құрылған айнымалылар. Олар **тек** сол функцияның ішінде қолжетімді және ол аяқталғаннан кейін жойылады.

```python
global_var = "Мен жаһандық айнымалымын"

def my_function():
    local_var = "Мен жергілікті айнымалымын"
    print(f"Функция ішінде: {local_var}")
    print(f"Функция ішінде жаһандықты оқи аламын: {global_var}")

my_function()

print(f"\nБағдарлама сыртында: {global_var}")
# Келесі жол қате тудырады, себебі local_var сыртта жоқ
# print(f"Бағдарлама сыртында: {local_var}") 
# NameError: name 'local_var' is not defined
```

**LEGB ережесі (Local, Enclosing, Global, Built-in):** Python айнымалыны алдымен жергілікті аймақтан, содан кейін "қоршаушы" (ішкі функциялар үшін), содан кейін жаһандық аймақтан және соңында кіріктірілген атаулар арасынан іздейді.

> **Маңызды:** Жаһандық айнымалыларды функция ішінде өзгертуден аулақ болыңыз. Деректерді параметрлер арқылы беріп, нәтижені `return` арқылы қайтарған дұрыс. Бұл кодты болжамды және жөндеуге оңай етеді.

## 5-бөлім. Функцияларды құжаттау (Docstrings)

Функцияларға құжаттама жолдарын (docstrings) қосу жақсы тәжірибе болып табылады. Бұл функция анықтамасынан кейін бірден келетін және оның не істейтінін, қандай аргументтер қабылдайтынын және не қайтаратынын сипаттайтын көпжолды түсініктеме (`"""..."""`).

```python
import math

def solve_quadratic_equation(a, b, c):
    """ax^2 + bx + c = 0 квадрат теңдеуінің түбірлерін есептейді.

    Args:
        a (int, float): x^2 жанындағы коэффициент.
        b (int, float): x жанындағы коэффициент.
        c (int, float): Бос мүше.

    Returns:
        tuple or None: Екі түбір, бір түбір бар кортеж немесе түбірлер жоқ болса, None.
    """
    if a == 0:
        return None # Бұл квадрат теңдеу емес

    discriminant = b**2 - 4*a*c

    if discriminant > 0:
        x1 = (-b + math.sqrt(discriminant)) / (2*a)
        x2 = (-b - math.sqrt(discriminant)) / (2*a)
        return (x1, x2)
    elif discriminant == 0:
        x = -b / (2*a)
        return (x,)
    else:
        return None

# Құжаттамаға осылай қол жеткізе аламыз:
help(solve_quadratic_equation)
```

## 6-бөлім. Тәжірибелік кейстер: Алғашқы функцияларымызды жазамыз

Теория жақсы, бірақ функцияларды түсінудің ең жақсы жолы — оларды жазу. Қазір жазып, іске қоса алатын бірнеше қарапайым мысалды қарастырайық.

### Кейс 1: Жолмен жұмыс істеуге арналған қарапайым функция

**Мәселе:** Адамның атын қабылдап, ресми сәлемдесуді қайтаратын функция жазу.

```python
def format_greeting(name):
    """Аты-жөні бар жолды қабылдап, сәлемдесуді қайтарады."""
    return f"Сәлеметсіз бе, {name}!"

# Функцияны шақырып, нәтижесін басып шығарамыз
greeting = format_greeting("Иван Петров")
print(greeting)

print(format_greeting("Мария Сидорова"))
```

**Мұнда не болып жатыр?**
1. Біз `name` деген бір параметрі бар `format_greeting` функциясын анықтадық.
2. Ішінде жаңа хабарлама құру үшін f-жолды қолдандық.
3. `return` осы жаңа хабарламаны функцияны шақырған жерге қайтарады.
4. Нәтижені айнымалыға (`greeting`) сақтауға немесе бірден басып шығаруға болады.

### Кейс 2: Есептеулері бар және сан қайтаратын функция

**Мәселе:** Сандар тізімінің арифметикалық ортасын есептейтін функция жазу.

```python
def calculate_average(numbers):
    """Сандар тізімінің орташа мәнін есептейді."""
    # Маңызды тексеріс: егер тізім бос болса ше?
    if not numbers:
        return 0 # Нөлге бөлуді болдырмау үшін 0 қайтарамыз
    
    total_sum = sum(numbers)
    count = len(numbers)
    average = total_sum / count
    return average

grades = [5, 4, 3, 5, 4, 4]
avg_grade = calculate_average(grades)
print(f"Орташа баға: {avg_grade:.2f}")

empty_list = []
avg_empty = calculate_average(empty_list)
print(f"Бос тізімнің орташа мәні: {avg_empty}")
```

**Мұнда не болып жатыр?**
1. Біз **шеткі жағдайды** (edge case) ескердік: егер тізім бос болса, оның ұзындығына (`len`) бөлуге болмайды. Функция 0 қайтарады.
2. Қарапайымдылық үшін кіріктірілген `sum()` және `len()` функцияларын қолдандық.
3. Функция ондық бөлшек (`float`) қайтарады, оны біз шығару кезінде форматтаймыз.

### Кейс 3: Бульдік мән (True/False) қайтаратын функция

**Мәселе:** Санның жұп екенін тексеретін функция жазу.

```python
def is_even(number):
    """Егер сан жұп болса True, әйтпесе False қайтарады."""
    return number % 2 == 0

# Мұндай функцияның нәтижесін шартты операторларда қолдану өте ыңғайлы
num1 = 10
if is_even(num1):
    print(f"{num1} саны — жұп.")
else:
    print(f"{num1} саны — тақ.")

num2 = 7
if is_even(num2):
    print(f"{num2} саны — жұп.")
else:
    print(f"{num2} саны — тақ.")
```

**Мұнда не болып жатыр?**
1. Біз `%` қалдықпен бөлу операторын қолдандық. Егер 2-ге бөлгендегі қалдық 0-ге тең болса, сан жұп.
2. `number % 2 == 0` өрнегінің өзі `True` немесе `False` мәнін береді. Біз осы нәтижені бірден қайтара (`return`) аламыз. Бұл өте кең таралған және ықшам тәсіл.
3. Біз предикат-функцияның (иә/жоқ деп жауап беретін) нәтижесі `if` конструкциясына қалай мінсіз сәйкес келетінін көрсеттік.

## Қорытынды

Функциялар — кез келген маңызды бағдарламаның негізгі құрылыс материалы. Олар келесі мүмкіндіктерді береді:
1.  **Кодты құрылымдау:** Үлкен тапсырмаларды кішкентай және түсінікті бөліктерге бөлу.

3.  **Оқылымдылықты жақсарту:** Код блоктарына мағыналы атаулар беру.
4.  **Жөндеуді жеңілдету:** Қатені үлкен скрипттен гөрі кішкентай, оқшауланған функциядан табу және түзету оңайырақ.
