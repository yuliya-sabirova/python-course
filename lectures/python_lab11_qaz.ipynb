### Жұмысқа аннотация

**Жұмыстың мақсаты:** Python-да сандық (жуық) және символдық (дәл) тәсілдердің айырмашылығын түсіне отырып, интегралдық есептеу міндеттерін шешуді үйрену. `SciPy` және `SymPy` кітапханаларының сәйкес құралдарын пайдаланып, анықталмаған, анықталған, меншіксіз және еселік интегралдарды шешуді меңгеру.

### Нұсқа нөмірін анықтау

Әр тапсырма үшін сіздің нұсқаңыз келесі формула бойынша есептеледі: **`(N - 1) % M + K`**, мұнда:
- `N` — сіздің топ тізіміндегі нөміріңіз.
- `M` — тапсырмадағы нұсқалар саны.
- `K` — тізімдегі бірінші нұсқаның нөмірі.
